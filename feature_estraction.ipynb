```python
# Assegnamo ad ogni paziente l'immagine corrispondende
def featureExtract(model, paths, t):
  features={}   #prepariamo il dataframe di output
  features=pd.DataFrame(features)
  for path in paths:
    #per ogni utente carichiamo e formattiamo le immagini
    if t=='x':
      img = keras.utils.load_img(path, target_size=(299, 299))
    if t=='e':
      img = keras.utils.load_img(path, target_size=(380, 380))
    x = keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature=model.predict(x)[0] #estraiamo le feature
    row=pd.DataFrame([feature])
    features=pd.concat([features,row], ignore_index=True)   #concateniamo tutto
  return features
```

```python
#eliminiamo l'ultimo strato di classificazione per estrarre solamente le feature
modelX=tf.keras.applications.xception.Xception(weights='imagenet',include_top=True)
modelX=Model(inputs=modelX.inputs, outputs=modelX.layers[-2].output)
```

```python
#eliminiamo l'ultimo strato di classificazione per estrarre solamente le feature
modelE=tf.keras.applications.EfficientNetB4(weights='imagenet',include_top=True)
modelE=Model(inputs=modelE.inputs, outputs=modelE.layers[-2].output)
```
